```
import os
import pandas as pd
from sqlalchemy import create_engine

file_path = 'HW_14_gutenberg.xlsx'
df = pd.read_excel(file_path)
df.columns = [col.lower() for col in df.columns]
df = df.where(pd.notnull(df), None)

database='postgres'
user = 'postgres'
host = 'localhost'
port = '5432'
db_password = os.getenv('PASSWORD_PG')

if not db_password:
    raise ValueError("Помилка: Змінна середовища PASSWORD_PG не знайдена! Додайте її командою (cmd) setx PASSWORD_PG 'ваш_пароль' ")


con_alch = create_engine('postgresql://{user}:{password}@{host}:{port}/{dbname}'.format(
dbname=database, user=user, password=db_password, host=host, port=port))
df.to_sql('books', con=con_alch, if_exists='replace', index=False)

```
